```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
font = {'weight' : 'normal',
        'size'   : 13}

plt.rc('font', **font)

plt.figure(figsize=(15,4))

plt.subplot(2,4,1)
x = np.linspace(0,0.5,5000)

y_mat = np.sin(100*x*2*np.pi)

plt.plot(x,y_mat, color='C0')

plt.xlim((0,0.02))
plt.ylim((-3,3))

plt.title('Fz - CAR')
y_ave = y_mat.mean(0)

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.yticks([])
plt.xticks([])

plt.subplot(2,4,5)

y_mat = np.sin(100*x*2*np.pi - np.pi)
plt.plot(x,y_mat, color='C1')


plt.xlim((0,0.02))
plt.ylim((-3,3))

plt.title('M1 - CAR')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.yticks([])
plt.xticks([])


ax = plt.subplot(2,4,(2,6))

ax.arrow(0,0,-0.5,0.5,head_width=0.1, head_length=0.1,ec = 'C0')
ax.arrow(0,0,0.5,-0.5,head_width=0.1, head_length=0.1,fc = 'C1',ec = 'C1')
ax.axvline(0,linestyle='--',color='k',alpha=0.4)
ax.axhline(0,linestyle='--',color='k',alpha=0.4)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.yticks([])
plt.xticks([])

plt.title('Fourier-transformed signal with CAR')
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.tight_layout()


# LMR

plt.subplot(2,4,3)
x = np.linspace(0,0.5,5000)

y_mat = np.sin(100*x*2*np.pi)

plt.plot(x,y_mat*2, color='C0')

plt.xlim((0,0.02))
plt.ylim((-3,3))

plt.title('Fz - M1')
y_ave = y_mat.mean(0)

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.yticks([])
plt.xticks([])

plt.subplot(2,4,7)

y_mat = np.sin(0*x*2*np.pi - np.pi)
plt.plot(x,y_mat, color='C1')


plt.xlim((0,0.02))
plt.ylim((-3,3))

plt.title('M1 - M1')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.yticks([])
plt.xticks([])


ax = plt.subplot(2,4,(4,8))

ax.arrow(0,0,-1,1,head_width=0.1, head_length=0.1,ec = 'C0')
ax.arrow(0,0,0,0,head_width=0.1, head_length=0.1,fc = 'C1',ec = 'C1')
ax.axvline(0,linestyle='--',color='k',alpha=0.4)
ax.axhline(0,linestyle='--',color='k',alpha=0.4)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.yticks([])
plt.xticks([])

plt.title('Fourier-transformed signal with LMR')
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.tight_layout()
```
