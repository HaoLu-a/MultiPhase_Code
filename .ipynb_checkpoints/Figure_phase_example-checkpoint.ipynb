```python
# import packages
import numpy as np
import mne
import matplotlib.pyplot as plt
import pickle
import pycircstat


from scipy.stats import circmean,circvar,ttest_rel
from mne.preprocessing import ICA
from scipy.io import loadmat, savemat
from mne.viz import plot_evoked_topo, plot_evoked
```

```python
font = {'weight' : 'normal',
        'size'   : 13}

plt.rc('font', **font)

plt.figure(figsize=(15,8))

plt.subplot(2,2,1)
x = np.linspace(0,0.5,512)
phase_shift = np.linspace(-0.5,0.5,8)

y_mat = np.empty((8,len(x)))
for i in range(8):
    y_mat[i,:] = np.sin(100*x*2*np.pi-phase_shift[i])

    plt.plot(x,y_mat[i,:])

plt.xlim((0,0.01))
plt.title('Example original signal')
y_ave = y_mat.mean(0)

plt.xlabel('Time(s)')
plt.ylabel('Amplitude')

plt.subplot(2,2,3)
x = np.linspace(0,0.5,512)

for i in range(8):
    plt.plot(x,y_mat[i,:]-y_ave)


plt.xlim((0,0.01))
plt.title('Example average-referenced signal')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')

ax = plt.subplot(2,2,(2,4))
freq = np.fft.rfftfreq(512,1/1024)

index = np.argmin(abs(freq-100))

xlab = [0,1]

phase_shift = [0.5,0.1,-0.1,-0.5]
for i in range(8):
    y1 = y_mat[i,:]
    y1_phase = [np.angle(np.fft.rfft(y1)[index]),
                         np.angle(np.fft.rfft(y1-y_ave)[index])]
    plt.plot(xlab,y1_phase,marker='o')

plt.xlim(-0.5,1.5)
plt.xticks([0,1])
ax.set_xticklabels(['Original','Average-referenced'])

plt.title('Phase shifted by average reference')
plt.ylabel('Phase')
plt.tight_layout()
```

```python
font = {'weight' : 'normal',
        'size'   : 13}

plt.rc('font', **font)

plt.figure(figsize=(8,6))


x = np.linspace(0,0.5,512)

phase_range = np.linspace(0,np.pi,10)

ave_amp = np.empty((len(phase_range)))
dif_amp = np.empty((len(phase_range)))
stds = np.empty((len(phase_range)))
for iPhase in range(len(phase_range)):
    phase_shift = np.linspace(-phase_range[iPhase],phase_range[iPhase],63)

    y_mat = np.empty((63,len(x)))
    for i in range(63):
        y_mat[i,:] = np.sin(100*x*2*np.pi-phase_shift[i])

    y_ave = y_mat.mean(0)
    stds[iPhase] = phase_shift.std()
    ave_amp[iPhase] = max(y_ave)
    dif_amp[iPhase] = max(y_mat[3,:]-y_ave)

    
plt.plot(stds,ave_amp, label = 'Amplitude of average reference')
plt.plot(stds,dif_amp, label = 'Amplitude of referenced EEG')
#plt.axhline(y=1, color='k', alpha=0.5, linestyle='--')

plt.axvline(x=0.8309348964303189, color = 'k', alpha = 0.5, linestyle='--')
plt.axvline(x=0.3448614696545039, color = 'k', alpha = 0.5, linestyle='--')

plt.ylim(0,1.2)
plt.xlim(0,1)
plt.xlabel('Standard deviation of equally-spaced phase')
plt.ylabel('Normalized amplitude')
plt.legend()
plt
```
