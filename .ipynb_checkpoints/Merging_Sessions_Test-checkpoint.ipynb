```python
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from random import sample
import pickle
import mne
from scipy.stats import circmean,circvar
from sklearn import decomposition
```

```python
# we want to check if the 34 Hz results from two sessions can be merged
# to test it, we may simply generate a new dataset with multiple electrodes

# the dataset will be with PLV ~ reference * location * Session

# first we compare the reference to reveal the difference
# just using PLV and with single channel and all channels

ID_list = [6,7,8,9,10,12,14,16,17,18,19,20,21,22]
Freq_list = [34.28, 91.42]
output = np.empty((len(ID_list)*2*len(Freq_list)*66,5))

# ID, reference(0 for ave, 1 for mastoid), frequency(43.43 and 98), channel name
# Fz, all chan * reference

row_num = 0

for iSubject in ID_list:# range(2,18):

    folder_path = "./jupyter_projects/MultiChannelSSSR/EEG_data_40_100_200Hz_LF/"
    sub_id = 'Sub' + str(iSubject) + '_LF'
    raw = mne.io.read_raw_bdf(folder_path + sub_id + '/' + sub_id + '_TMR0_MSK.bdf',
                                   stim_channel='Status', preload=True, exclude = ['EXG3','EXG4','EXG5','EXG6','EXG7','EXG8'])
    events = mne.find_events(raw)

    epochs = mne.Epochs(raw,events=events, tmin=-0.5, tmax=0.5)
    epochs.load_data()
    ep_m = epochs.copy().set_eeg_reference(ref_channels=['EXG1','EXG2'])
    ep_a = epochs.copy().set_eeg_reference(ref_channels='average')
    
    for iRef in range(2):
        if iRef==0:
            data = ep_a.get_data()[:,:,int(np.round((0.5)*4096)):int(np.round((0.5+0.43752)*4096))]
        else:
            data = ep_m.get_data()[:,:,int(np.round((0.5)*4096)):int(np.round((0.5+0.43752)*4096))]
            

        xw = np.fft.rfft(np.hamming(data.shape[2])*data, axis=2)
        plv = abs((xw/abs(xw)).mean(0))
        
        freq = np.fft.rfftfreq(data.shape[2],1/4096)
        
        out = xw[:,0:66,:]
        sample_dat = np.concatenate((np.real(out),np.imag(out)),axis=1)
            
        for iFreq in range(len(Freq_list)):
            index = np.argmin(abs(freq-Freq_list[iFreq]))
            
            for iChannel in range(66):
                output[row_num,0] = iSubject
                output[row_num,1] = iRef
                output[row_num,2] = Freq_list[iFreq]
                output[row_num,3] = iChannel
                output[row_num,4] = plv[iChannel,index] 

                row_num +=1

output_L = output

file_handle = open(folder_path + '/allchannel_PLV_LF.obj','wb')
pickle.dump(output, file_handle)
file_handle.close()
```

```python
# load saved location plv

folder_path = "./jupyter_projects/MultiChannelSSSR/EEG_data_40_100_200Hz_LF/"
file_handle = open(folder_path +'/allchannel_PLV_LF.obj','rb')
output_L = pickle.load(file_handle)
file_handle.close()

folder_path = "./jupyter_projects/MultiChannelSSSR/EEG_data_40_100_200Hz_HF/"
file_handle = open(folder_path +'/allchannel_PLV_HF.obj','rb')
output_H = pickle.load(file_handle)
file_handle.close()
```

```python
# combine matrix L and H for statistical test with R

output = output_L
Freq_set = {34.28, 91.42}

# add column name and generate data frame
label = [output[iE,2] in Freq_set for iE in range(output.shape[0])]
df = pd.DataFrame(output[label,0:6],columns=['ID','ref','freq','channel','PLV'])

# label the session as low freq
df['Session'] = 'Low'
df_L = df
df_L.head()

output = output_H
Freq_set = {34.28, 217.13}


# add column name and generate data frame
label = [output[iE,2] in Freq_set for iE in range(output.shape[0])]
df = pd.DataFrame(output[label,0:6],columns=['ID','ref','freq','channel','PLV'])

# label the session as High freq
df['Session'] = 'High'
df_H = df
df_H.head()

# concatenate dataframes
df_All = df_L.append(df_H)
df_All.to_csv(r'./jupyter_projects/MultiChannelSSSR/Location_PLV_Both_Session.csv')
```
