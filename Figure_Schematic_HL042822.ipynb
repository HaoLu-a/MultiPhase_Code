```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import string
```

```python
figure_path = './jupyter_projects/MultiChannelSSSR/Figures/'
img1 = mpimg.imread(figure_path + 'BioRenderExport.jpeg')
```

```python
font = {'weight' : 'normal',
        'size'   : 13}

plt.rc('font', **font)

plt.figure(figsize=(15,8))

plt.subplot(1,2,1)

plt.imshow(img1)
plt.axis('off')
plt.text(0, 0.1, string.ascii_uppercase[0], 
            size=30, weight='bold')

plt.subplot(4,4,3)
x = np.linspace(0,0.5,5000)

y_mat = np.sin(100*x*2*np.pi)

plt.plot(x,y_mat, color='C0')

plt.xlim((0,0.02))
plt.ylim((-3,3))

plt.title('Cz - CAR')
y_ave = y_mat.mean(0)

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.yticks([])
plt.xticks([])

plt.text(-0.005, -0.1, string.ascii_uppercase[1], 
            size=30, weight='bold')

plt.subplot(4,4,7)

y_mat = np.sin(100*x*2*np.pi - np.pi)
plt.plot(x,y_mat, color='C1')


plt.xlim((0,0.02))
plt.ylim((-3,3))

plt.title('M1 - CAR')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.yticks([])
plt.xticks([])


ax = plt.subplot(4,4,(4,8))

ax.arrow(0,0,-0.5,0.5,head_width=0.1, head_length=0.1,ec = 'C0')
ax.arrow(0,0,0.5,-0.5,head_width=0.1, head_length=0.1,fc = 'C1',ec = 'C1')
ax.axvline(0,linestyle='--',color='k',alpha=0.4)
ax.axhline(0,linestyle='--',color='k',alpha=0.4)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.yticks([])
plt.xticks([])

plt.title('Fourier-transformed signal with CAR')
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.tight_layout()


# LMR

plt.subplot(4,4,11)
x = np.linspace(0,0.5,5000)

y_mat = np.sin(100*x*2*np.pi)

plt.plot(x,y_mat*2, color='C0')

plt.xlim((0,0.02))
plt.ylim((-3,3))

plt.title('Cz - M1')
y_ave = y_mat.mean(0)

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.yticks([])
plt.xticks([])

plt.subplot(4,4,15)

y_mat = np.sin(0*x*2*np.pi - np.pi)
plt.plot(x,y_mat, color='C1')


plt.xlim((0,0.02))
plt.ylim((-3,3))

plt.title('M1 - M1')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.yticks([])
plt.xticks([])


ax = plt.subplot(4,4,(12,16))

ax.arrow(0,0,-1,1,head_width=0.1, head_length=0.1,ec = 'C0')
ax.arrow(0,0,0,0,head_width=0.1, head_length=0.1,fc = 'C1',ec = 'C1')
ax.axvline(0,linestyle='--',color='k',alpha=0.4)
ax.axhline(0,linestyle='--',color='k',alpha=0.4)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.yticks([])
plt.xticks([])

plt.title('Fourier-transformed signal with LMR')
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.tight_layout()

plt.savefig(figure_path+'Figure6.jpg', dpi=300)
```
